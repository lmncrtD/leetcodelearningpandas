```python
### 177. Nth Highest Salary

import pandas as pd

data = [[1, 100], [2, 200], [3, 300]]
employee = pd.DataFrame(data, columns=['id', 'salary']).astype({'id':'Int64', 'salary':'Int64'})
```

```python
def nth_highest_salary(employee: pd.DataFrame, N: int) -> pd.DataFrame:
    '''
    Требуется вернуть N-е по величине максимальное уникальное значение.
    '''
    if employee['salary'].drop_duplicates().count() < N or N <= 0:
        return pd.DataFrame({f'getNthHighestSalary({N})': [pd.NA]})
    else:
        salary_unique = employee.sort_values(by='salary').drop_duplicates(subset='salary')
        return pd.DataFrame({f'getNthHighestSalary({N})': [salary_unique.iloc[-N]['salary']]})
```

```python
nth_highest_salary(employee, 1)
```

```python
nth_highest_salary(employee, -1)
```

```python
nth_highest_salary(employee, 0)
```

```python
nth_highest_salary(employee, 4)
```

```python
nth_highest_salary(employee, 2)
```

```python
nth_highest_salary(employee, 3)
```
